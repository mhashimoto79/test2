```python
path = '../tests/data/4_MApps/Layout.xlsx'
df_layout = pd.read_excel(path,skiprows=1)
path = '../tests/data/4_MApps/Rawdata.xlsx'
df_raw = pd.read_excel(path)
```

```python
#df_raw.head()
```

```python
#layoutデータ加工
df_layout['質問番号'].ffill(inplace=True)
df_layout['質問タイプ'].ffill(inplace=True)
```

```python
# labelデータ作成
df_label = df_raw

for i in range(0, len(df_raw.columns)):
    col_name = df_raw.columns[i]
    df_tmp = df_layout[df_layout['ラベル'] == col_name]
    #質問タイプがMAの場合
    if df_tmp['質問タイプ'].values[0] == 'M':
        label = df_tmp['質問文／選択肢'].values[0]
        df_label[col_name] = df_raw[col_name].replace(1, label).replace(0, None)
    elif df_tmp['質問タイプ'].values[0] == 'MTS' or df_tmp['質問タイプ'].values[0] == 'S' :
        qid = df_tmp['質問番号'].values[0]
        df_tmp = df_layout[df_layout['質問番号'] == qid]
        df_tmp['ラベル'].ffill(inplace=True)
        df_tmp = df_tmp[df_tmp['ラベル'] == col_name]
        df_tmp['選択肢番号'] = df_tmp['選択肢番号'].astype(pd.Int64Dtype(), errors='ignore')
        df_tmp_dict = df_tmp.set_index('選択肢番号')['質問文／選択肢'].to_dict()
        df_label[col_name] = df_raw[col_name].map(df_tmp_dict)
        
```

```python
df_label.to_excel('./tmp.xlsx')
```

```python
#ラベルデータ加工：横持ちデータ、縦持ちデータ変換
# 縦持ちデータ変換対象のカラム名を取得
col_list = df_layout[df_layout['質問タイプ'].isin(['M','MTS'])]['ラベル'].drop_duplicates().dropna().tolist()
```

```python
#横持ち（Drayage）、縦持ち（Vertical Holding）
df_label_drayage = df_label.drop(columns=col_list)
df_label_vertical = df_label[col_list]
```

```python
df_label_vertical = df_label_vertical.stack().reset_index()
df_label_vertical = df_label_vertical.rename(columns={'level_0':"ID",'level_1':'Q',0:'A'})
df_tmp = df_label[['MID']]
df_label_vertical = pd.merge(df_label['MID'],df_label_vertical, left_index=True, right_on='ID')
df_label_vertical = df_label_vertical.drop(columns=['ID'])
```
